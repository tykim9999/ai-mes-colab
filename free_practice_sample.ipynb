## 1단계: 데이터 생성

아래 템플릿에서 `___` 부분을 여러분 업종에 맞게 채워서 Colab AI에 입력하세요.

```

```

**우리 업종이면 이렇게:**

| 업종 | X (공정 조건) | Y (품질 결과) |
|------|--------------|--------------|
| 정밀가공 | 절삭속도, 이송률, 절삭깊이, 공구마모, 냉각유량 | 표면조도, 치수편차, 가공시간 |
| 사출 성형 | 사출압, 보압, 금형온도, 냉각시간, 수지온도 | 치수정밀도, 수축률, 외관등급 |
| 식품/바이오 | 배양온도, pH, 배양시간, 교반속도, 원료농도 | 생균수, 안정성, 수율 |
| 카드/전자 | 인쇄속도, 접합온도, 접합압력, 경화시간, 습도 | 접착강도, 외관불량률, 내구성 |
| 금속/방화문 | 절단속도, 용접전류, 용접시간, 판재두께, 냉각시간 | 인장강도, 변형량, 기밀도 |
| 펌프/설비 | 회전속도, 유입압력, 임펠러직경, 간극, 유체점도 | 유량, 효율, 진동값 |

Y: 표면조도(0.1-6.0), 치수편차(0.5-30um), 가공시간(10-300초).
